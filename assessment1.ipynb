```python
del crime['Total']
crime.head()
```

```python
crime.groupby((crime.index.year//10)*10).Murder.sum()
```

```python
tmp = crime[['Violent','Property','Murder','Forcible_Rape','Robbery','Aggravated_assault','Burglary','Larceny_Theft','Vehicle_Theft']]
tmp.groupby((tmp.index.year//10)*10).sum()
```
